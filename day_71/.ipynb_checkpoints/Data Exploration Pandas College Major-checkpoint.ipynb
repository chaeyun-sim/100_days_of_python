# Challenge

## 1. 중간 경력 봉급이 가장 높은 대학 전공은? 이 전공 졸업생들은 얼마를 벌까?

```python
midCareers_major = clean_df['Undergraduate Major'].loc[clean_df['Mid-Career Median Salary'].idxmax()]
clean_df[clean_df['Undergraduate Major'] == midCareers_major]
```

## 2. 초봉이 가장 낮은 대학의 전공은 무엇이며 졸업 후 졸업생은 얼마를 벌까?

```python
starting_major = clean_df['Undergraduate Major'].loc[clean_df['Starting Median Salary'].idxmin()]
clean_df[clean_df['Undergraduate Major'] == starting_major]
```


## 3. 중간 경력 봉급이 가장 낮은 대학의 전공은 무엇이며 이 학위로 얼마나 벌 수 있을까?

```python
midCareers_min_major = clean_df['Undergraduate Major'].loc[clean_df['Mid-Career Median Salary'].idxmin()]
clean_df[clean_df['Undergraduate Major'] == midCareers_min_major]
```

## Lowest-risk Major

```python
clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
# clean_df['Mid-Career 90th Percentile Salary'].subtract(clean_df['Mid-Career 10th Percentile Salary'])
```

```python
spread_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1, 'Spread', spread_col)
clean_df.head()
```

```python
low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
```

# Challenge

## 1. .sort_values()를 사용하여 잠재력이 가장 높은 학위를 찾을 수 있을까? 90번쨰 백분위수에서 가장 높은 값을 가진 5개의 학위를 찾아보자

```python
df_90th = clean_df.sort_values(by='Mid-Career 90th Percentile Salary', ascending=False)
df_90th[['Mid-Career 90th Percentile Salary', 'Undergraduate Major']].head()
```

## 2. 급여에서 가장 큰 분포를 보이는 학위를 찾을 수 있을까? 졸업 후 고소득자와 저소득자 사이 소득차이가 가장 큰 전공은 무엇일까?
